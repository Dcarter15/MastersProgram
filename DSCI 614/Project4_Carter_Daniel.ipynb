```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import HashingVectorizer
```

```python
twitter_df = pd.read_csv('/content/Twitter_Data.csv')
twitter_df.head()
```

```python
#check for NaN values)
twitter_df.isnull().sum()
```

```python
#fill NaN values with blank string and 0
twitter_df['clean_text'].fillna('', inplace=True)
twitter_df['category'].fillna(0, inplace=True)
```

```python
#verify no more NaN
twitter_df.isnull().sum()
```

### Convert column of clean text to matrix of tokens using CountVectorizer and unigrams and bigrams

```python
vectorizer = CountVectorizer()
matrix = vectorizer.fit_transform(twitter_df['clean_text'])
```

```python
vectorizer = CountVectorizer(ngram_range=(1,2))
matrix = vectorizer.fit_transform(twitter_df['clean_text'])
```

### Perform tf-idf analysis on clean text using CountVectorizer and TfidfTransformer

```python
vectorizer = CountVectorizer()
matrix = vectorizer.fit_transform(twitter_df['clean_text'])

tf_idf_transformer = TfidfTransformer(use_idf=True, smooth_idf=True, sublinear_tf=False)
tf_idf_matrix = tf_idf_transformer.fit_transform(matrix)
```

### Perform tf-idf analysis on clean text using Tfidfvectorizer

```python
tfidf_vectorizer = TfidfVectorizer(use_idf=True, smooth_idf=True, sublinear_tf=False)
tfidf_matrix = tfidf_vectorizer.fit_transform(twitter_df['clean_text'])
```

### Perform tf-idf on clean text using HashingVectorizer and TfidfTransformer

```python
hash_vectorizer = HashingVectorizer()
hash_matrix = hash_vectorizer.fit_transform(twitter_df['clean_text'])
tf_idf_transformer = TfidfTransformer(use_idf=True, smooth_idf=True, sublinear_tf=False)
tf_idf_matrix = tf_idf_transformer.fit_transform(hash_matrix)
```
