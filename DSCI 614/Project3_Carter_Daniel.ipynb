```python
import pandas as pd
import numpy as np
import spacy
import matplotlib.pyplot as plt
import PyPDF2
import re
```

### Extract text

```python
pdf = '/content/tax-efficient-withdrawal-strategies.pdf'
pdfFile = open(pdf, 'rb')
pdfReader = PyPDF2.PdfReader(pdfFile)

#extract text from each page in the PDF
text = []
for page in pdfReader.pages:
    text.append(page.extract_text())

alltext = ' '.join(text)

#remove \n and punctuation from text
alltext = re.sub(r'\n', ' ', alltext)
alltext = re.sub(r'[^\w\s]', '', alltext)

pdfFile.close()
```

### Extract all tokens and lemma

```python
nlp = spacy.load('en_core_web_sm')
doc = nlp(alltext)
#print out tokens
tokens = [token.text for token in doc]
print(tokens)
```

```python
#print out lemma
for token in doc:
  print(f'Text: {token.text}; Lemma: {token.lemma_}')
```

### Remove all stop words

```python
stop_words = nlp.Defaults.stop_words
no_stop_words = [token.text for token in doc if token.text not in stop_words]
print(no_stop_words)
```

### Customize stop words. Add tax and account and remove full

```python
add_stop_words = ['tax', 'account']
for token in add_stop_words:
  nlp.Defaults.stop_words.add(token)
for token in add_stop_words:
  nlp.vocab[token].is_stop = False

remove_stop_words = ['full']
for token in remove_stop_words:
  nlp.Defaults.stop_words.remove(token)
for token in remove_stop_words:
  nlp.vocab[token].is_stop = False
```

```python
tokens_no_stop = [token.text for token in doc if token.text not in nlp.Defaults.stop_words]
print(tokens_no_stop)
```

### Speech tagging

```python
for token in doc:
  print(f'Text: {token.text}; POS: {token.pos_}')
```

### Visualize the dependancy

```python
spacy.displacy.render(doc, style='dep', jupyter=True)
```

### Preform Named entities recognition

```python
for ent in doc.ents:
  print(f'Text: {ent.text}; Label: {ent.label_}')
```

### Visualize the MONEY and CARDINAL entities

```python
options = {'ents': ['MONEY', 'CARDINAL']}
spacy.displacy.render(doc, style='ent', jupyter=True, options=options)
```
