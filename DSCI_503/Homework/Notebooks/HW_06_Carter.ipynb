## Problem 7: Batting Data - Batting Averages

```python
batting_avg = batting.filter(items=['teamID', 'yearID', 'H', 'AB']).groupby(['teamID', 'yearID']).sum().reset_index()
batting_avg['BA'] = batting_avg['H']/batting_avg['AB']
batting_avg = batting_avg.sort_values('BA', ascending=False)
batting_avg.head(10)
```

## Problem 8: Batting Data - Cardinals vs Cubs

```python
stl_batting = batting[['yearID', 'H', 'AB', 'HR']].loc[(batting['teamID']=='SLN') & (batting['yearID'] >= 1900)].groupby('yearID').sum().sort_values('yearID').reset_index()
stl_batting['BA'] = stl_batting['H']/stl_batting['AB']
chi_batting = batting[['yearID', 'H', 'AB', 'HR']].loc[(batting['teamID']=='CHN') & (batting['yearID'] >= 1900)].groupby('yearID').sum().sort_values('yearID').reset_index()
chi_batting['BA'] = chi_batting['H']/chi_batting['AB']
```

```python
plt.figure(figsize=[12,4])
plt.subplot(2,2,1)
plt.plot(stl_batting['yearID'], stl_batting['BA'], color='red', label='Cardinals')
plt.plot(chi_batting['yearID'], chi_batting['BA'], color='blue', label='Cubs')
plt.xlabel('Year')
plt.ylabel('Batting Average')
plt.title('Batting Average By Year')
plt.legend()
plt.show()

plt.subplot(2,2,2)
plt.plot(stl_batting['yearID'], stl_batting['HR'], color='red', label='Cardinals')
plt.plot(chi_batting['yearID'], chi_batting['HR'], color='blue', label='Cubs')
plt.xlabel('Year')
plt.ylabel('Batting Average')
plt.title('Home Runs By Year')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
stl_higher_ba = np.round(np.mean(np.where(stl_batting['BA'] > chi_batting['BA'], 1,0)),4)
stl_higher_hr = np.round(np.mean(np.where(stl_batting['HR'] > chi_batting['HR'], 1,0)),4)
print(stl_higher_ba)
print(stl_higher_hr)
```
