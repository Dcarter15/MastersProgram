```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mpg_df = pd.read_csv('mpg.csv')
mpg_df.head()
```

# Histogram

```python
plt.figure(figsize=(12,6))
year1 = mpg_df[mpg_df['year']==1999]['displ']
year2 = mpg_df[mpg_df['year']==2008]['displ']
labels = ['1999', '2008']
plt.hist([year1,year2], 30, stacked=True, label=labels)
plt.legend()
plt.title('Stacked Histogram of enigne displacement colored by year')
plt.xlabel('Displacement')
plt.ylabel('Frequency')
plt.show()
```

# Violin 


```python
plt.figure(figsize=[12,6])
sns.violinplot(x='year', y='hwy', data=mpg_df, scale='width', inner='quartile')
plt.title('Violinplot of highway mpg by year')
```

# Scatter

```python
plt.figure(figsize=[15,8])
sns.scatterplot(mpg_df['cty'], mpg_df['hwy'], hue=mpg_df['year'], palette=['red', 'blue'])
sns.regplot(mpg_df['cty'], mpg_df['hwy'], scatter=False, color='black')
plt.legend()
plt.xlim(0,40)
plt.title('Scatter plot city vs highway mpg with regression line')
```
